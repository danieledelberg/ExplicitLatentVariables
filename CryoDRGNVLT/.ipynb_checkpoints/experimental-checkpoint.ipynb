```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(2)
```

```python
pca.fit(z_old)
pc = pca.transform(z_old)
plt.scatter(pc[:,0], pc[:,1], alpha=0.1, s=0.1)
for i in range(10):
    plt.text(pc[i, 0], pc[i, 1], i)
    plt.scatter(pc[i, 0], pc[i, 1], s=10, c='red')
plt.show()
```

```python
np.random.seed(0)
perturb = np.random.permutation(pc)
```

```python
inverse_perturb = pca.inverse_transform(perturb)
inverse_pc = pca.transform(inverse_perturb)
plt.scatter(inverse_pc[:,0], inverse_pc[:,1], alpha=0.1, s=0.1)
for i in range(10):
    plt.text(inverse_pc[i, 0], inverse_pc[i, 1], i)
    plt.scatter(inverse_pc[i, 0], inverse_pc[i, 1], s=10, c='red')
plt.show()
```

```python
from cryodrgn import analysis
import numpy as np
import pickle
import matplotlib.pyplot as plt

for i in np.linspace(0,99,10):
    with open(f'datav2/zero_01_b00/z_vals.{int(i)}.pkl','rb') as f:
        z, z_logvar = pickle.load(f)
    pc, pca = analysis.run_pca(z)
    print(int(i))
    plt.scatter(pc[:,0], pc[:,1], alpha=0.1, s=0.1)
    plt.scatter(pc[0:10,0], pc[0:10,1], s=4, color='red')
    for i in range(10):
        plt.text(pc[i,0], pc[i,1], str(i), size=12)
        
    # Include this for beta=1000
    plt.scatter(pc[41056, 0], pc[41056,1], s=4, color='red')
    plt.text(pc[41056, 0], pc[41056,1], str(41056), size=12)
    plt.show()
```
